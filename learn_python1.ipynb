```python
try:
    stop = True
    while stop:

        if stop == 0:
            stop = False

        input_number_x = float(input("Input x"))
        input_number_y = float(input("Input y"))

        input_action = input("Input action")

        if input_action == '+':
            input_summery = input_number_x + input_number_y
            print("Summary: '+' ", input_summery)
        elif input_action == '-':
            input_summery = input_number_x - input_number_y
            print("NOT Summary: '-' ", input_summery)
        elif input_action == '*':
            input_summery = input_number_x * input_number_y
            print("Ymnosh: '*' ", input_summery)
        elif input_action == '/':
            input_summery = input_number_x / input_number_y
            print("Delite: '/' ", input_summery)
        else:
            print('not correct action')

        stop = int(input("if give leave write 0"))
except ValueError:
    print("Error: ", input_action, " not action")
```
