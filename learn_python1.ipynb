```python
import numpy as np
import os
import random as rd
from random import randint
```

```python
def calculate():
    while True:
        try:
            x = float(input("Введите число X: "))
            y = float(input("Введите число Y: "))
            action = input("Введите действие (+, -, *, /): ").strip()

            if action == '+':
                result = x + y
            elif action == '-':
                result = x - y
            elif action == '*':
                result = x * y
            elif action == '/':
                result = x / y
            else:
                print("Ошибка: неизвестное действие.")
                continue

            print(f"Результат: {x} {action} {y} = {result}")

        except ValueError:
            print("Ошибка: введите корректные числа!")
        except ZeroDivisionError:
            print("Ошибка: деление на ноль!")

        stop = input("Чтобы выйти, введите 0 (или нажмите Enter для продолжения): ")
        if stop.strip() == '0':
            print("Работа завершена. Спасибо!")
            break

calculate()
```

```python
def user_random_element(kolich, diaposon_a, diaposon_b):
    user_list = []
    for i in range(kolich):
        random_element = rd.randint(diaposon_a, diaposon_b)
        user_list.append(random_element)
    return user_list

def delete_list(user_list):
    user_list = sorted(set(user_list))
    return user_list

try:
    kolich = int(input("Количество элементов: "))
    diaposon_a = int(input("Начало диапазона: "))
    diaposon_b = int(input("Конец диапазона: "))

    if diaposon_a > diaposon_b:
        print("Ошибка: начало диапазона не может быть больше конца!")
    else:
        user_list = user_random_element(kolich, diaposon_a, diaposon_b)
        print("Исходный список:", user_list)

        user_list = delete_list(user_list)
        print("Список без повторов:", user_list)

except ValueError:
    print("Ошибка: введите корректные числа!")
```
